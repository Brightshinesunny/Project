```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# 샘플 데이터 생성
np.random.seed(42)
data = pd.DataFrame({
    "constant": np.random.normal(0.8, 0.1, 100),
    "educ": np.random.normal(0.08, 0.01, 100),
    "pexp": np.random.normal(0.08, 0.02, 100),
    "broken_home": np.random.normal(0.0, 0.1, 100)
})

# 초기 설정
sns.set(style="white", context="talk")
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

# 각 축 설정
ax1, ax2, ax3, ax4 = ax.ravel()

# 애니메이션 초기화 함수
def init():
    # 축 초기화
    ax1.clear()
    ax2.clear()
    ax3.clear()
    ax4.clear()

    # 각 축에 타이틀 추가
    ax1.set_title('constant vs educ')
    ax2.set_title('constant vs pexp')
    ax3.set_title('constant vs broken_home')
    ax4.set_title('educ vs pexp')

# 애니메이션 업데이트 함수
def animate(i):
    # 각 축을 지우고 새로운 데이터를 그립니다.
    ax1.clear()
    ax2.clear()
    ax3.clear()
    ax4.clear()

    # i번째 데이터까지 플롯에 추가
    sns.scatterplot(x=data['constant'][:i], y=data['educ'][:i], ax=ax1, color='blue')
    sns.scatterplot(x=data['constant'][:i], y=data['pexp'][:i], ax=ax2, color='orange')
    sns.scatterplot(x=data['constant'][:i], y=data['broken_home'][:i], ax=ax3, color='green')
    sns.scatterplot(x=data['educ'][:i], y=data['pexp'][:i], ax=ax4, color='red')

    # 각 플롯에 제목 다시 추가
    ax1.set_title('constant vs educ')
    ax2.set_title('constant vs pexp')
    ax3.set_title('constant vs broken_home')
    ax4.set_title('educ vs pexp')

# 애니메이션 객체 생성
anim = FuncAnimation(fig, animate, init_func=init, frames=len(data), interval=100, blit=False)

# 애니메이션을 파일로 저장
anim.save('/content/animation.mp4', fps=10, extra_args=['-vcodec', 'libx264'])

# Colab에서 파일 재생을 위한 HTML 코드 생성
from IPython.display import HTML
from base64 import b64encode

mp4 = open('/content/animation.mp4','rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML(f'<video width=400 controls><source src="{data_url}" type="video/mp4"></video>')
```
