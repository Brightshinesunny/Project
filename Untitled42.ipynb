```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
import matplotlib.patches as patches

# Given Fisher matrix from context
fisher_matrix = np.array([
    [2.86182e11, -3.93346e10, -1.13964e10, 4090.32, -30.2739, -1.2571e10, -4.34567e7],
    [-3.93346e10, 7.17572e9, 2.06522e9, -2535.59, 34.6039, 4.39014e9, 9.21181e6],
    [-1.13964e10, 2.06522e9, 5.94667e8, -775.863, 11.0671, 1.28862e9, 2.65332e6],
    [4090.32, -2535.59, -775.863, 0.038369, -0.00105898, -19.548, -7.00417],
    [-30.2739, 34.6039, 11.0671, -0.00105898, 3.0888e-6, 474.957, 0.124442],
    [-1.2571e10, 4.39014e9, 1.28862e9, -19.548, 474.957, 1.32662e10, 8.58071e6],
    [-4.34567e7, 9.21181e6, 2.65332e6, -7.00417, 0.124442, 8.58071e6, 12996]
])

# Calculate the covariance matrix
cov_matrix = la.inv(fisher_matrix)

# True values for each parameter
true_values = np.array([5.9932e-6, 0.2222, 0.01, 237, 101, 0, 0])
param_names = [r'$M_c$', r'$\eta$', r'$\chi_1$', r'$\Lambda$', r'$\delta \Lambda$', r'$t_c$', r'$\phi_c$']
num_params = len(param_names)

# Calculate bias as the square root of variances from the diagonal of the covariance matrix
bias_values = np.sqrt(np.diag(cov_matrix))
confidence_levels = [1.0, 2.0, 3.0]  # for 1, 2, 3 intervals

# Set up plot with adjusted labels for readability
fig, axes = plt.subplots(num_params, num_params, figsize=(15, 15))
fig.subplots_adjust(hspace=0.3, wspace=0.3)

for i in range(num_params):
    for j in range(num_params):
        ax = axes[i, j]

        if i == j:
            # Diagonal: plot marginal distributions centered on true values with bias marked
            mu = true_values[i]
            sigma = np.sqrt(cov_matrix[i, i])
            bias = bias_values[i]
            x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
            ax.plot(x, np.exp(-0.5 * ((x - mu) / sigma) ** 2), color="gray")
            ax.axvline(mu, color="red", linestyle="--", label="True Value")  # True value
            ax.axvline(mu + bias, color="blue", linestyle="--", label="Bias")  # Bias offset
            ax.set_yticks([])
            ax.set_title(f"{param_names[i]} = {mu:.2e} (Bias: {bias:.1e})", fontsize=10)

            # Display true value and bias on the x-axis with shorter notation
            ax.set_xticks([mu, mu + bias])
            ax.set_xticklabels([f"{mu:.2e}", f"{(mu + bias):.2e}"])
            ax.get_xticklabels()[0].set_color("red")
            ax.get_xticklabels()[1].set_color("blue")

        elif i > j:
            # Off-diagonal: plot 2D Gaussian contours centered on true values with bias offset lines
            cov_submatrix = cov_matrix[np.ix_([j, i], [j, i])]
            eigenvalues, eigenvectors = la.eigh(cov_submatrix)
            angle = np.degrees(np.arctan2(*eigenvectors[:, 0][::-1]))

            for level in confidence_levels:
                width, height = 2 * level * np.sqrt(eigenvalues)
                ellipse = patches.Ellipse((true_values[j], true_values[i]), width, height,
                                          angle=angle, fill=False, edgecolor="gray", linestyle="--")
                ax.add_patch(ellipse)

            ax.set_xlim(true_values[j] - 3 * np.sqrt(cov_matrix[j, j]), true_values[j] + 3 * np.sqrt(cov_matrix[j, j]))
            ax.set_ylim(true_values[i] - 3 * np.sqrt(cov_matrix[i, i]), true_values[i] + 3 * np.sqrt(cov_matrix[i, i]))
            ax.axvline(true_values[j], color="red", linestyle="--")  # True value
            ax.axhline(true_values[i], color="red", linestyle="--")  # True value

            # Display true values and bias on x and y axes with shorter notation
            ax.axvline(true_values[j] + bias_values[j], color="blue", linestyle="--")  # Bias offset x-axis
            ax.axhline(true_values[i] + bias_values[i], color="blue", linestyle="--")  # Bias offset y-axis
            ax.set_xticks([true_values[j], true_values[j] + bias_values[j]])
            ax.set_yticks([true_values[i], true_values[i] + bias_values[i]])
            ax.get_xticklabels()[0].set_color("red")
            ax.get_xticklabels()[1].set_color("blue")
            ax.get_yticklabels()[0].set_color("red")
            ax.get_yticklabels()[1].set_color("blue")

        else:
            ax.axis('off')

# Add legend for bias and true values
fig.legend(loc="upper right", fontsize=10)

plt.show()
```
