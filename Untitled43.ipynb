```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
import corner

# Given Fisher matrix from context
fisher_matrix = np.array([
    [2.86182e11, -3.93346e10, -1.13964e10, 4090.32, -30.2739, -1.2571e10, -4.34567e7],
    [-3.93346e10, 7.17572e9, 2.06522e9, -2535.59, 34.6039, 4.39014e9, 9.21181e6],
    [-1.13964e10, 2.06522e9, 5.94667e8, -775.863, 11.0671, 1.28862e9, 2.65332e6],
    [4090.32, -2535.59, -775.863, 0.038369, -0.00105898, -19.548, -7.00417],
    [-30.2739, 34.6039, 11.0671, -0.00105898, 3.0888e-6, 474.957, 0.124442],
    [-1.2571e10, 4.39014e9, 1.28862e9, -19.548, 474.957, 1.32662e10, 8.58071e6],
    [-4.34567e7, 9.21181e6, 2.65332e6, -7.00417, 0.124442, 8.58071e6, 12996]
])

# Calculate the covariance matrix
cov_matrix = la.inv(fisher_matrix)

# True values and parameter names
true_values = np.array([5.9932e-6, 0.2222, 0.01, 237, 101, 0, 0])
param_names = [r'$M_c$', r'$\eta$', r'$\chi_1$', r'$\Lambda$', r'$\delta \Lambda$', r'$t_c$', r'$ \phi_c$']
num_params = len(param_names)  # Define the number of parameters

# Generate random samples based on the covariance matrix
num_samples = 10000
samples = np.random.multivariate_normal(mean=true_values, cov=cov_matrix, size=num_samples)

# Calculate the mean (estimated central values) and bias
estimated_means = np.mean(samples, axis=0)
biases = estimated_means - true_values

# Create the corner plot
fig = corner.corner(samples, labels=param_names, truths=true_values,
                    truth_color="red", show_titles=True, title_fmt=".2e")

# Add bias lines and annotations
for i, ax in enumerate(fig.get_axes()):
    # Only add bias lines to 1D histograms
    if i % (num_params + 1) == 0:
        param_index = i // (num_params + 1)
        ax.axvline(estimated_means[param_index], color="blue", linestyle="--", label="Bias")
        ax.text(0.9, 0.9, f"Bias: {biases[param_index]:.2e}",
                transform=ax.transAxes, color="blue", ha="right")

plt.legend(loc="upper right")
plt.show()
```
