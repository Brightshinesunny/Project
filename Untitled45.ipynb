```python
import numpy as np
import matplotlib.pyplot as plt

# Covariance matrices from the provided data
cov_with_prior = np.array([
    [8.43686e-7, -7.25569e-7, 0.000282998, -0.0705624, 4.98656],
    [-7.25569e-7, 0.00632714, -0.0246069, 7.45001, -430.198],
    [0.000282998, -0.0246069, 0.095761, -27.467, 1683.31],
    [-0.0705624, 7.45001, -27.467, 246609, 4.26454e6],
    [4.98656, -430.198, 1683.31, 4.26454e6, 1.38716e8]
])

cov_no_prior = np.array([
    [3.38308e-6, -3.34059e-4, 0.00125343, -7.74022, -124.938],
    [-3.34059e-4, 0.0332797, -0.124605, 801.557, 13033.1],
    [0.00125343, -0.124605, 0.466789, -2970.18, -48196.7],
    [-7.74022, 801.557, -2970.18, 2.41953e7, 4.11992e8],
    [-124.938, 13033.1, -48196.7, 4.11992e8, 7.08541e9]
])

# Solar mass and updated True Values
solar_mass = 4.9257e-6
m1 = 2 * solar_mass
m2 = 1 * solar_mass
M = m1 + m2
eta = (m1 * m2) / (M ** 2)
Mc = M * (eta ** (3 / 5))

true_values_updated = [Mc, eta, 0.025, 237, 101]  # True values reflecting solar mass
labels_updated = [r"$M_c$", r"$\eta$", r"$\chi$", r"$\tilde{\lambda}$", r"$\delta\tilde{\lambda}$"]

# Plotting parameters
fig, axes = plt.subplots(len(labels_updated), len(labels_updated), figsize=(15, 15))
levels = [1, 2, 3]  # 1, 2, 3 confidence levels
colors_no_prior = ['blue', 'orange', 'green']
colors_with_prior = ['cyan', 'magenta', 'yellow']

# Helper function for plotting confidence ellipses
def plot_confidence_ellipses(ax, cov, mean, idx1, idx2, colors, levels):
    """Plot confidence ellipses with updated true values."""
    cov_sub = cov[np.ix_([idx1, idx2], [idx1, idx2])]
    eigvals, eigvecs = np.linalg.eigh(cov_sub)
    angle = np.degrees(np.arctan2(eigvecs[1, 0], eigvecs[0, 0]))
    for n_std, color in zip(levels, colors):
        width, height = 2 * n_std * np.sqrt(eigvals)
        ell = plt.matplotlib.patches.Ellipse(xy=(mean[idx1], mean[idx2]), width=width, height=height,
                                             angle=angle, edgecolor=color, facecolor='none')
        ax.add_patch(ell)
    # Adjust axis limits based on covariance scale
    std_x = np.sqrt(cov[idx1, idx1])
    std_y = np.sqrt(cov[idx2, idx2])
    ax.set_xlim(mean[idx1] - 5 * std_x, mean[idx1] + 5 * std_x)
    ax.set_ylim(mean[idx2] - 5 * std_y, mean[idx2] + 5 * std_y)
    ax.set_xlabel(labels_updated[idx1], fontsize=10)
    ax.set_ylabel(labels_updated[idx2], fontsize=10)
    # Mark true value
    ax.plot(mean[idx1], mean[idx2], 'r*', markersize=10)
    ax.text(mean[idx1], mean[idx2], f"({mean[idx1]:.2e}, {mean[idx2]:.2e})", color='red', fontsize=8)

# Plot corner plot
for i in range(len(labels_updated)):
    for j in range(len(labels_updated)):
        ax = axes[i, j]
        if j > i:
            ax.axis('off')
            continue
        elif i == j:
            # Diagonal: Histogram of the distributions
            std_no_prior = np.sqrt(cov_no_prior[i, i])
            std_with_prior = np.sqrt(cov_with_prior[i, i])
            ax.hist(np.random.normal(true_values_updated[i], std_no_prior, 1000), bins=30, density=True,
                    alpha=0.6, label="No Prior", color='blue')
            ax.hist(np.random.normal(true_values_updated[i], std_with_prior, 1000), bins=30, density=True,
                    alpha=0.6, label="With Prior", color='red')
            ax.set_xlim(true_values_updated[i] - 5 * std_no_prior, true_values_updated[i] + 5 * std_no_prior)
            # Mark True Value
            ax.axvline(true_values_updated[i], color='red', linestyle='--', label=f"True: {true_values_updated[i]:.2e}")
        else:
            # Off-diagonal: Confidence ellipses
            plot_confidence_ellipses(ax, cov_no_prior, np.array(true_values_updated), i, j,
                                     colors_no_prior, levels)
            plot_confidence_ellipses(ax, cov_with_prior, np.array(true_values_updated), i, j,
                                     colors_with_prior, levels)
        ax.tick_params(axis='both', which='major', labelsize=8)

# Add legend to the figure
handles = [
    plt.matplotlib.patches.Patch(color='blue', label='No Prior: Marginalized posterior distribution (1, 2, 3)'),
    plt.matplotlib.patches.Patch(color='cyan', label='With Prior: Marginalized posterior distribution (1, 2, 3)'),
    plt.Line2D([0], [0], color='red', linestyle='--', label='True Value (red line and star)')
]
fig.legend(handles=handles, loc='upper right', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#  
solar_mass = 4.9257e-6

# True Values  ( )
true_values_solar = [
    0.9873 * solar_mass,  # Mc scaled by solar mass
    0.2222,              # eta (dimensionless, remains unchanged)
    0.025,               # chi (dimensionless, remains unchanged)
    237,                 # lambda_tilde (unchanged)
    101                  # delta_lambda_tilde (unchanged)
]

# Covariance Matrix (Prior-less)  ( )
cov_no_prior = np.array([
    [3.38308e-6, -3.34059e-4, 0.00125343, -7.74022, -124.938],
    [-3.34059e-4, 0.0332797, -0.124605, 801.557, 13033.1],
    [0.00125343, -0.124605, 0.466789, -2970.18, -48196.7],
    [-7.74022, 801.557, -2970.18, 2.41953e7, 4.11992e8],
    [-124.938, 13033.1, -48196.7, 4.11992e8, 7.08541e9],
])

# Covariance Matrix (With Prior)  ( )
cov_with_prior = np.array([
    [8.43686e-7, -7.25569e-7, 0.000282998, -0.0705624, 4.98656],
    [-7.25569e-7, 0.00632714, -0.0246069, 7.45001, -430.198],
    [0.000282998, -0.0246069, 0.095761, -27.467, 1683.31],
    [-0.0705624, 7.45001, -27.467, 246609, 4.26454e6],
    [4.98656, -430.198, 1683.31, 4.26454e6, 1.38716e8],
])

#   (Prior-less)
parameters_no_prior_solar_sample = np.random.multivariate_normal(
    mean=true_values_solar, cov=cov_no_prior_solar, size=2000
)

#   (With Prior)
parameters_with_prior_solar_sample = np.random.multivariate_normal(
    mean=true_values_solar, cov=cov_with_prior_solar, size=2000
)

#  
labels_trimmed = [r"$M_c$", r"$\eta$", r"$\chi$", r"$\tilde{\lambda}$", r"$\delta\tilde{\lambda}$"]

#   
def plot_all_pairs(data, labels, true_values):
    sns.set(style="whitegrid")
    num_vars = len(labels)
    fig, axes = plt.subplots(num_vars, num_vars, figsize=(20, 20))
    for i in range(num_vars):
        for j in range(num_vars):
            if i < j:
                axes[i, j].axis('off')  # No plot for upper triangle
            elif i == j:
                # Marginal histograms
                sns.histplot(data[:, i], kde=True, color="blue", ax=axes[i, j])
                axes[i, j].axvline(true_values[i], color='red', linestyle='--', label=f'True: {true_values[i]:.2e}')
                axes[i, j].legend()
                axes[i, j].set_xlabel(labels[i])
                axes[i, j].set_ylabel('Density')
            else:
                # 2D KDE plots with confidence ellipses
                sns.kdeplot(
                    x=data[:, j], y=data[:, i], fill=True, cmap="Blues", levels=6, ax=axes[i, j]
                )
                axes[i, j].plot(true_values[j], true_values[i], 'r*', markersize=10, label='True Value')
                axes[i, j].legend()
                axes[i, j].set_xlabel(labels[j])
                axes[i, j].set_ylabel(labels[i])
    plt.tight_layout()
    plt.show()

# Prior     
print("Plotting Prior-less Covariance Matrix with Solar Mass Scaling...")
plot_all_pairs(parameters_no_prior_solar_sample, labels_trimmed, true_values_solar)

# Prior     
print("Plotting With Prior Covariance Matrix with Solar Mass Scaling...")
plot_all_pairs(parameters_with_prior_solar_sample, labels_trimmed, true_values_solar)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import multivariate_normal

# Define constants
solarmass = 4.9257e-6

# Updated true values with solar mass
true_values = {
    "M_c": 0.9873 * solarmass,
    "eta": 0.2222,
    "chi": 0.025,
    "tilde_lambda": 237,
    "delta_tilde_lambda": 101,
    "t_c": 0,
    "phi_c": 0,
}

# Labels for the plot
param_labels = [
    r"$M_c$",
    r"$\eta$",
    r"$\chi$",
    r"$\tilde{\lambda}$",
    r"$\delta\tilde{\lambda}$",
    r"$t_c$",
    r"$\phi_c$",
]

# True values in vector form
true_values_vector = np.array([
    true_values["M_c"],
    true_values["eta"],
    true_values["chi"],
    true_values["tilde_lambda"],
    true_values["delta_tilde_lambda"],
    true_values["t_c"],
    true_values["phi_c"],
])

# Covariance matrices (No Prior and With Prior)
cov_no_prior = np.array([
    [3.38308e-6, -3.34059e-4, 1.25343e-3, -7.74022, -124.938, -1.23081e-6, -7.87847e-3],
    [-3.34059e-4, 3.32797e-2, -1.24605e-1, 801.557, 13033.1, 1.26557e-3, 0.725551],
    [1.25343e-3, -1.24605e-1, 0.466789, -2970.18, -48196.7, -4.69918e-3, -2.77338],
    [-7.74022, 801.557, -2970.18, 2.41953e7, 4.11992e8, 36.2562, 10068.7],
    [-124.938, 13033.1, -48196.7, 4.11992e8, 7.08541e9, 610.046, 139634],
    [-1.23081e-6, 1.26557e-3, -4.69918e-3, 36.2562, 610.046, 0.000551194, 0.0183046],
    [-7.87847e-3, 0.725551, -2.77338, 10068.7, 139634, 0.0183046, 28.9305],
])

cov_with_prior = np.array([
    [8.43686e-7, -7.25569e-7, 2.82998e-4, -7.05624e-2, 4.98656, -7.44897e-7, -4.25171e-3],
    [-7.25569e-7, 6.32714e-3, -2.46069e-2, 7.45001, -430.198, 6.94968e-4, 0.357259],
    [2.82998e-4, -2.46069e-2, 0.095761, -27.467, 1683.31, -2.65784e-4, -1.40157],
    [-7.05624e-2, 7.45001, -27.467, 2.46609e5, 4.26454e6, 0.362716, 60.3981],
    [4.98656, -430.198, 1683.31, 4.26454e6, 1.38716e8, -0.50787, -27451.7],
    [-7.44897e-7, 6.94968e-4, -2.65784e-4, 0.362716, -0.50787, 1.2676e-6, 0.0029592],
    [-4.25171e-3, 0.357259, -1.40157, 60.3981, -27451.7, 0.0029592, 22.6108],
])

# Function to plot marginalized posterior distribution
def plot_marginalized_posterior(cov_matrix, title):
    fig, axes = plt.subplots(len(param_labels), len(param_labels), figsize=(15, 15))
    for i in range(len(param_labels)):
        for j in range(len(param_labels)):
            ax = axes[i, j]
            if i == j:
                # Marginal distributions (diagonal)
                mean, var = true_values_vector[i], cov_matrix[i, i]
                x = np.linspace(mean - 3 * np.sqrt(var), mean + 3 * np.sqrt(var), 500)
                y = multivariate_normal.pdf(x, mean=mean, cov=var)
                ax.plot(x, y, color="blue")
                ax.fill_between(x, y, color="blue", alpha=0.3)
            elif i > j:
                # Joint distributions (lower triangle)
                mean = [true_values_vector[j], true_values_vector[i]]
                cov = cov_matrix[np.ix_([j, i], [j, i])]
                x, y = np.mgrid[
                    mean[0] - 3 * np.sqrt(cov[0, 0]): mean[0] + 3 * np.sqrt(cov[0, 0]): 100j,
                    mean[1] - 3 * np.sqrt(cov[1, 1]): mean[1] + 3 * np.sqrt(cov[1, 1]): 100j,
                ]
                pos = np.dstack((x, y))
                rv = multivariate_normal(mean, cov)
                ax.contourf(x, y, rv.pdf(pos), levels=10, cmap="Blues", alpha=0.7)
            else:
                ax.axis("off")
            if i == len(param_labels) - 1:
                ax.set_xlabel(param_labels[j])
            if j == 0:
                ax.set_ylabel(param_labels[i])
    fig.suptitle(title, fontsize=16)
    plt.tight_layout()
    plt.show()

# Plot marginalized posterior for both cases
plot_marginalized_posterior(cov_no_prior, "Marginalized Posterior Without Prior (1, 2, 3)")
plot_marginalized_posterior(cov_with_prior, "Marginalized Posterior With Prior (1, 2, 3)")
```
