```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# \tilde{\Lambda}와 \delta\Lambda의 정의
def tilde_lambda(m1, m2, lambda1, lambda2):
    return (16 / 13) * (
        ((m1 + 12 * m2) * m1**4 * lambda1 + (m2 + 12 * m1) * m2**4 * lambda2) /
        (m1 + m2)**5
    )

def delta_lambda(lambda1, lambda2):
    return 0.5 * (lambda1 - lambda2)

# 범위 및 샘플 수 설정
num_samples = 100000
m1_samples = np.random.uniform(0.5, 2.5, num_samples)
m2_samples = np.random.uniform(0.5, 2.5, num_samples)
lambda1_samples = np.random.uniform(0, 5000, num_samples)
lambda2_samples = np.random.uniform(0, 5000, num_samples)

# \tilde{\Lambda}와 \delta\Lambda 샘플 생성
tilde_lambda_samples = tilde_lambda(m1_samples, m2_samples, lambda1_samples, lambda2_samples)
delta_lambda_samples = delta_lambda(lambda1_samples, lambda2_samples)

# 결합 데이터 생성
data = np.vstack([tilde_lambda_samples, delta_lambda_samples])

# 커널 밀도 추정
kde = gaussian_kde(data)

# 그리드 설정
xmin, xmax = tilde_lambda_samples.min(), tilde_lambda_samples.max()
ymin, ymax = delta_lambda_samples.min(), delta_lambda_samples.max()
x, y = np.linspace(xmin, xmax, 100), np.linspace(ymin, ymax, 100)
X, Y = np.meshgrid(x, y)
positions = np.vstack([X.ravel(), Y.ravel()])

# 밀도 계산
Z = np.reshape(kde(positions).T, X.shape)

# 시각화
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, levels=50, cmap='rainbow')
plt.colorbar(label='Density')
plt.xlabel(r'$\tilde{\Lambda}$')
plt.ylabel(r'$\delta\Lambda$')
plt.title('Joint Prior Distribution of $\tilde{\Lambda}$ and $\delta\Lambda$')
plt.show()
```
