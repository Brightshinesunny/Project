```python
# 분리된 밀도 분포를 각각 플롯하기 위해, \tilde{\Lambda}와 \delta\Lambda의 1차원 KDE를 계산합니다.
from scipy.stats import gaussian_kde
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# \tilde{\Lambda}와 \delta\Lambda의 정의
def tilde_lambda(m1, m2, lambda1, lambda2):
    return (16 / 13) * (
        ((m1 + 12 * m2) * m1**4 * lambda1 + (m2 + 12 * m1) * m2**4 * lambda2) /
        (m1 + m2)**5
    )

def delta_lambda(lambda1, lambda2):
    return 0.5 * (lambda1 - lambda2)

# 범위 및 샘플 수 설정
num_samples = 100000
m1_samples = np.random.uniform(0.5, 2.5, num_samples)
m2_samples = np.random.uniform(0.5, 2.5, num_samples)
lambda1_samples = np.random.uniform(0, 5000, num_samples)
lambda2_samples = np.random.uniform(0, 5000, num_samples)

# \tilde{\Lambda}와 \delta\Lambda 샘플 생성
tilde_lambda_samples = tilde_lambda(m1_samples, m2_samples, lambda1_samples, lambda2_samples)
delta_lambda_samples = delta_lambda(lambda1_samples, lambda2_samples)
# 개별 KDE 계산
kde_tilde_lambda = gaussian_kde(tilde_lambda_samples)
kde_delta_lambda = gaussian_kde(delta_lambda_samples)

# x축 범위 설정
tilde_lambda_range = np.linspace(tilde_lambda_samples.min(), tilde_lambda_samples.max(), 1000)
delta_lambda_range = np.linspace(delta_lambda_samples.min(), delta_lambda_samples.max(), 1000)

# PDF 계산
tilde_lambda_density = kde_tilde_lambda(tilde_lambda_range)
delta_lambda_density = kde_delta_lambda(delta_lambda_range)

# 개별 분포 시각화
plt.figure(figsize=(14, 6))

# \tilde{\Lambda} 분포
plt.subplot(1, 2, 1)
plt.plot(tilde_lambda_range, tilde_lambda_density, label=r'$\tilde{\Lambda}$ Density')
plt.fill_between(tilde_lambda_range, tilde_lambda_density, alpha=0.5)
plt.xlabel(r'$\tilde{\Lambda}$')
plt.ylabel('Density')
plt.title(r'Distribution of $\tilde{\Lambda}$')
plt.legend()

# \delta\Lambda 분포
plt.subplot(1, 2, 2)
plt.plot(delta_lambda_range, delta_lambda_density, label=r'$\delta\Lambda$ Density', color='orange')
plt.fill_between(delta_lambda_range, delta_lambda_density, alpha=0.5, color='orange')
plt.xlabel(r'$\delta\Lambda$')
plt.ylabel('Density')
plt.title(r'Distribution of $\delta\Lambda$')
plt.legend()

plt.tight_layout()
plt.show()
```
