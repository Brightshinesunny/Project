```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Step 1: Define given true values and parameter names
true_values = np.array([8.33202e-6, 0.248521, 0.01, 20.5, 4.8, 0, 0])
param_names = [r"$M_c$", r"$\eta$", r"$\chi_1$", r"$\Lambda$", r"$\delta \Lambda$", r"$t_c$", r"$\phi_c$"]

# Step 2: Define covariance matrices provided by the user
cov_matrix_prior_user = np.array([
    [3.0341e-7, -0.0000222726, 0.000106704, -0.149605, -10.1403, -4.10809e-7, -0.00100227],
    [-0.0000222726, 0.00165516, -0.00791195, 12.1292, 833.872, 0.0000322314, 0.0720583],
    [0.000106704, -0.00791195, 0.0378365, -56.9595, -3904.09, -0.000152401, -0.346713],
    [-0.149605, 12.1292, -56.9595, 175.948, 1.34175e7, 0.366091, 376.301],
    [-10.1403, 833.872, -3904.09, 1.34175e7, 1.03865e9, 26.9674, 24025.9],
    [-4.10809e-7, 0.0000322314, -0.000152401, 0.366091, 26.9674, 8.27812e-7, 0.00115853],
    [-0.00100227, 0.0720583, -0.346713, 376.301, 24025.9, 0.00115853, 3.45927]
])

cov_matrix_no_prior_user = np.array([
    [6.59206e-7, -0.0000509403, 0.000241494, -0.551501, -40.6821, -1.25521e-6, -0.00191613],
    [-0.0000509403, 0.00396552, -0.0187743, 44.5582, 3298.6, 0.000100342, 0.145637],
    [0.000241494, -0.0187743, 0.0889077, -209.391, -15489.2, -0.000472575, -0.692714],
    [-0.551501, 44.5582, -209.391, 634.261, 4.82757e7, 1.32679, 1403.59],
    [-40.6821, 3298.6, -15489.2, 4.82757e7, 3.69007e9, 100.021, 102061.0],
    [-1.25521e-6, 0.000100342, -0.000472575, 1.32679, 100.021, 2.84274e-6, 0.00331947],
    [-0.00191613, 0.145637, -0.692714, 1403.59, 102061.0, 0.00331947, 5.81214]
])

# Step 3: Generate samples based on these covariance matrices
np.random.seed(42)  # For reproducibility
samples_prior_user = np.random.multivariate_normal(true_values, cov_matrix_prior_user, size=5000)
samples_no_prior_user = np.random.multivariate_normal(true_values, cov_matrix_no_prior_user, size=5000)

# Step 4: Plotting with 1-sigma ellipses and KDE
n_params = len(param_names)
fig, axes = plt.subplots(n_params, n_params, figsize=(15, 15))

for i in range(n_params):
    for j in range(n_params):
        ax = axes[i, j]
        if i == j:  # Diagonal: 1D KDE plots
            kde_no_prior_user = gaussian_kde(samples_no_prior_user[:, i])
            kde_prior_user = gaussian_kde(samples_prior_user[:, i])
            x_vals = np.linspace(min(samples_no_prior_user[:, i].min(), samples_prior_user[:, i].min()),
                                 max(samples_no_prior_user[:, i].max(), samples_prior_user[:, i].max()), 500)
            ax.plot(x_vals, kde_no_prior_user(x_vals), color='blue', label='No Prior', lw=1.5)
            ax.plot(x_vals, kde_prior_user(x_vals), color='green', label='With Prior', lw=1.5)
            ax.axvline(true_values[i], color='red', linestyle='--', label='True Value')
            ax.set_title(param_names[i], fontsize=10)
        elif i > j:  # Lower triangle: 1-sigma ellipses only
            x_vals = np.linspace(min(samples_no_prior_user[:, j].min(), samples_prior_user[:, j].min()),
                                 max(samples_no_prior_user[:, j].max(), samples_prior_user[:, j].max()), 100)
            y_vals = np.linspace(min(samples_no_prior_user[:, i].min(), samples_prior_user[:, i].min()),
                                 max(samples_no_prior_user[:, i].max(), samples_prior_user[:, i].max()), 100)
            X, Y = np.meshgrid(x_vals, y_vals)
            kde_no_prior_user = gaussian_kde(np.vstack([samples_no_prior_user[:, j], samples_no_prior_user[:, i]]))
            Z_no_prior_user = kde_no_prior_user(np.vstack([X.ravel(), Y.ravel()])).reshape(X.shape)
            kde_prior_user = gaussian_kde(np.vstack([samples_prior_user[:, j], samples_prior_user[:, i]]))
            Z_prior_user = kde_prior_user(np.vstack([X.ravel(), Y.ravel()])).reshape(X.shape)
            ax.contour(X, Y, Z_no_prior_user, colors='blue', linestyles='solid', levels=[np.max(Z_no_prior_user) * 0.606], alpha=0.7)
            ax.contour(X, Y, Z_prior_user, colors='green', linestyles='solid', levels=[np.max(Z_prior_user) * 0.606], alpha=0.7)
            ax.axvline(true_values[j], color='red', linestyle='--', lw=1)
            ax.axhline(true_values[i], color='red', linestyle='--', lw=1)
        else:  # Upper triangle: blank
            ax.axis('off')
        if j == 0:
            ax.set_ylabel(param_names[i], fontsize=8)
        if i == n_params - 1:
            ax.set_xlabel(param_names[j], fontsize=8)

axes[0, 0].legend(loc='upper right', fontsize=6, frameon=False)
plt.tight_layout()
plt.show()
```

```python
# True values and samples
true_values = np.array([8.33202e-6, 0.248521, 0.01, 20.5, 4.8, 0, 0])  # 참값
samples_prior_user_mean = np.mean(samples_prior_user, axis=0)  # 프라이어가 있는 샘플의 평균
samples_no_prior_user_mean = np.mean(samples_no_prior_user, axis=0)  # 프라이어가 없는 샘플의 평균

# Bias calculation
bias_prior = samples_prior_user_mean - true_values  # 프라이어가 있는 경우의 바이어스
bias_no_prior = samples_no_prior_user_mean - true_values  # 프라이어가 없는 경우의 바이어스

# Display results
print("Bias with Prior:", bias_prior)
print("Bias without Prior:", bias_no_prior)
```

```python
# 모든 플롯 중심을 참값으로 맞추고, 그래프 설명(legend)을 겹치지 않게 수정한 코드

# 타원의 해상도를 높이고 플롯 중심을 참값으로 설정
fig, axes = plt.subplots(n_params, n_params, figsize=(15, 15))

for i in range(n_params):
    for j in range(n_params):
        ax = axes[i, j]
        if i == j:  # 대각선: 1D KDE 플롯
            kde_no_prior_user = gaussian_kde(samples_no_prior_user[:, i])
            kde_prior_user = gaussian_kde(samples_prior_user[:, i])
            x_vals = np.linspace(true_values[i] - 3 * np.std(samples_no_prior_user[:, i]),
                                 true_values[i] + 3 * np.std(samples_no_prior_user[:, i]), 500)
            ax.plot(x_vals, kde_no_prior_user(x_vals), color='blue', label='No Prior', lw=1.5)
            ax.plot(x_vals, kde_prior_user(x_vals), color='green', label='With Prior', lw=1.5)
            ax.axvline(true_values[i], color='red', linestyle='--', label='True Value')  # 참값 표시
            ax.axvline(samples_no_prior_mean[i], color='blue', linestyle=':', label='No Prior Mean')  # 바이어스 위치
            ax.axvline(samples_prior_mean[i], color='green', linestyle=':', label='Prior Mean')  # 바이어스 위치
            ax.set_xlim(true_values[i] - 3 * np.std(samples_no_prior_user[:, i]),
                        true_values[i] + 3 * np.std(samples_no_prior_user[:, i]))
            ax.set_title(param_names[i], fontsize=10, pad=10)
        elif i > j:  # 하단 삼각형: 1시그마 타원만 표시
            x_vals = np.linspace(true_values[j] - 3 * np.std(samples_no_prior_user[:, j]),
                                 true_values[j] + 3 * np.std(samples_no_prior_user[:, j]), 200)
            y_vals = np.linspace(true_values[i] - 3 * np.std(samples_no_prior_user[:, i]),
                                 true_values[i] + 3 * np.std(samples_no_prior_user[:, i]), 200)
            X, Y = np.meshgrid(x_vals, y_vals)

            kde_no_prior_user = gaussian_kde(np.vstack([samples_no_prior_user[:, j], samples_no_prior_user[:, i]]))
            Z_no_prior_user = kde_no_prior_user(np.vstack([X.ravel(), Y.ravel()])).reshape(X.shape)

            kde_prior_user = gaussian_kde(np.vstack([samples_prior_user[:, j], samples_prior_user[:, i]]))
            Z_prior_user = kde_prior_user(np.vstack([X.ravel(), Y.ravel()])).reshape(X.shape)

            # 1시그마 레벨만 표시 (최고 밀도의 60.6%)
            ax.contour(X, Y, Z_no_prior_user, colors='blue', linestyles='solid', levels=[np.max(Z_no_prior_user) * 0.606], alpha=0.7)
            ax.contour(X, Y, Z_prior_user, colors='green', linestyles='solid', levels=[np.max(Z_prior_user) * 0.606], alpha=0.7)

            # 바이어스 위치를 점으로 추가
            ax.scatter(samples_no_prior_mean[j], samples_no_prior_mean[i], color='blue', s=40, label='No Prior Mean')
            ax.scatter(samples_prior_mean[j], samples_prior_mean[i], color='green', s=40, label='Prior Mean')
            ax.axvline(true_values[j], color='red', linestyle='--', lw=1)
            ax.axhline(true_values[i], color='red', linestyle='--', lw=1)
            ax.set_xlim(true_values[j] - 3 * np.std(samples_no_prior_user[:, j]),
                        true_values[j] + 3 * np.std(samples_no_prior_user[:, j]))
            ax.set_ylim(true_values[i] - 3 * np.std(samples_no_prior_user[:, i]),
                        true_values[i] + 3 * np.std(samples_no_prior_user[:, i]))
        else:  # 상단 삼각형: 공백 처리
            ax.axis('off')

        # 라벨 추가
        if j == 0:
            ax.set_ylabel(param_names[i], fontsize=8, labelpad=10)
        if i == n_params - 1:
            ax.set_xlabel(param_names[j], fontsize=8, labelpad=10)

# 범례를 플롯 외부로 설정
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', fontsize=10, ncol=4, frameon=False, bbox_to_anchor=(0.5, 1.05))

plt.tight_layout()
plt.show()
```
