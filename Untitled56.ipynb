```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# 1. **참값과 파라미터 이름 정의**
true_values = np.array([8.33202e-6, 0.248521, 0.01, 20.5, 4.8, 0, 0])
param_names = [r"$M_c$", r"$\eta$", r"$\chi_1$", r"$\Lambda$", r"$\delta \Lambda$", r"$t_c$", r"$\phi_c$"]

# 2. **공분산 행렬 (프라이어 있는 경우와 없는 경우)**
cov_matrix_prior_user = np.array([
    [3.0341e-7, -0.0000222726, 0.000106704, -0.149605, -10.1403, -4.10809e-7, -0.00100227],
    [-0.0000222726, 0.00165516, -0.00791195, 12.1292, 833.872, 0.0000322314, 0.0720583],
    [0.000106704, -0.00791195, 0.0378365, -56.9595, -3904.09, -0.000152401, -0.346713],
    [-0.149605, 12.1292, -56.9595, 175.948, 1.34175e7, 0.366091, 376.301],
    [-10.1403, 833.872, -3904.09, 1.34175e7, 1.03865e9, 26.9674, 24025.9],
    [-4.10809e-7, 0.0000322314, -0.000152401, 0.366091, 26.9674, 8.27812e-7, 0.00115853],
    [-0.00100227, 0.0720583, -0.346713, 376.301, 24025.9, 0.00115853, 3.45927]
])

cov_matrix_no_prior_user = np.array([
    [6.59206e-7, -0.0000509403, 0.000241494, -0.551501, -40.6821, -1.25521e-6, -0.00191613],
    [-0.0000509403, 0.00396552, -0.0187743, 44.5582, 3298.6, 0.000100342, 0.145637],
    [0.000241494, -0.0187743, 0.0889077, -209.391, -15489.2, -0.000472575, -0.692714],
    [-0.551501, 44.5582, -209.391, 634.261, 4.82757e7, 1.32679, 1403.59],
    [-40.6821, 3298.6, -15489.2, 4.82757e7, 3.69007e9, 100.021, 102061.0],
    [-1.25521e-6, 0.000100342, -0.000472575, 1.32679, 100.021, 2.84274e-6, 0.00331947],
    [-0.00191613, 0.145637, -0.692714, 1403.59, 102061.0, 0.00331947, 5.81214]
])

# 3. **샘플 생성**
np.random.seed(42)
samples_prior_user = np.random.multivariate_normal(true_values, cov_matrix_prior_user, size=5000)
samples_no_prior_user = np.random.multivariate_normal(true_values, cov_matrix_no_prior_user, size=5000)

# 4. **평균과 중앙값 계산**
samples_prior_mean = np.mean(samples_prior_user, axis=0)
samples_no_prior_mean = np.mean(samples_no_prior_user, axis=0)
samples_prior_median = np.median(samples_prior_user, axis=0)
samples_no_prior_median = np.median(samples_no_prior_user, axis=0)

# 5. **바이어스 계산**
bias_prior_mean = samples_prior_mean - true_values
bias_no_prior_mean = samples_no_prior_mean - true_values
bias_prior_median = samples_prior_median - true_values
bias_no_prior_median = samples_no_prior_median - true_values

# 6. **플롯 생성**
n_params = len(param_names)
fig, axes = plt.subplots(n_params, n_params, figsize=(15, 15))

for i in range(n_params):
    for j in range(n_params):
        ax = axes[i, j]
        if i == j:  # Diagonal: 1D KDE plots
            kde_no_prior_user = gaussian_kde(samples_no_prior_user[:, i])
            kde_prior_user = gaussian_kde(samples_prior_user[:, i])
            x_vals = np.linspace(true_values[i] - 3 * np.std(samples_no_prior_user[:, i]),
                                 true_values[i] + 3 * np.std(samples_no_prior_user[:, i]), 300)
            ax.plot(x_vals, kde_no_prior_user(x_vals), color='blue', label='No Prior KDE', lw=1.5)
            ax.plot(x_vals, kde_prior_user(x_vals), color='green', label='With Prior KDE', lw=1.5)
            ax.axvline(true_values[i], color='red', linestyle='--', label='True Value')
            ax.axvline(samples_no_prior_mean[i], color='blue', linestyle='-', label=f'Mean Bias (No Prior: {bias_no_prior_mean[i]:.2e})')
            ax.axvline(samples_prior_mean[i], color='green', linestyle='-', label=f'Mean Bias (Prior: {bias_prior_mean[i]:.2e})')
            ax.axvline(samples_no_prior_median[i], color='blue', linestyle=':', label=f'Median Bias (No Prior: {bias_no_prior_median[i]:.2e})')
            ax.axvline(samples_prior_median[i], color='green', linestyle=':', label=f'Median Bias (Prior: {bias_prior_median[i]:.2e})')
            ax.set_xlim(true_values[i] - 3 * np.std(samples_no_prior_user[:, i]),
                        true_values[i] + 3 * np.std(samples_no_prior_user[:, i]))
            ax.set_title(f"{param_names[i]}\nMean Bias: {bias_prior_mean[i]:.2e}, Median Bias: {bias_prior_median[i]:.2e}", fontsize=10, pad=10)
        elif i > j:  # Lower triangle: 1-sigma ellipses
            reduced_samples_no_prior = samples_no_prior_user[:1000, [j, i]]
            reduced_samples_prior = samples_prior_user[:1000, [j, i]]

            kde_no_prior_user = gaussian_kde(reduced_samples_no_prior.T)
            kde_prior_user = gaussian_kde(reduced_samples_prior.T)

            x_vals = np.linspace(true_values[j] - 3 * np.std(samples_no_prior_user[:, j]),
                                 true_values[j] + 3 * np.std(samples_no_prior_user[:, j]), 100)
            y_vals = np.linspace(true_values[i] - 3 * np.std(samples_no_prior_user[:, i]),
                                 true_values[i] + 3 * np.std(samples_no_prior_user[:, i]), 100)
            X, Y = np.meshgrid(x_vals, y_vals)

            Z_no_prior_user = kde_no_prior_user(np.vstack([X.ravel(), Y.ravel()])).reshape(X.shape)
            Z_prior_user = kde_prior_user(np.vstack([X.ravel(), Y.ravel()])).reshape(X.shape)

            ax.contour(X, Y, Z_no_prior_user, colors='blue', linestyles='solid', levels=[np.max(Z_no_prior_user) * 0.606], alpha=0.7)
            ax.contour(X, Y, Z_prior_user, colors='green', linestyles='solid', levels=[np.max(Z_prior_user) * 0.606], alpha=0.7)

            ax.scatter(samples_no_prior_mean[j], samples_no_prior_mean[i], color='blue', s=40, label='Mean (No Prior)')
            ax.scatter(samples_prior_mean[j], samples_prior_mean[i], color='green', s=40, label='Mean (Prior)')
            ax.scatter(samples_no_prior_median[j], samples_no_prior_median[i], color='blue', s=40, marker='x', label='Median (No Prior)')
            ax.scatter(samples_prior_median[j], samples_prior_median[i], color='green', s=40, marker='x', label='Median (Prior)')
            ax.axvline(true_values[j], color='red', linestyle='--', lw=1)
            ax.axhline(true_values[i], color='red', linestyle='--', lw=1)
            ax.set_xlim(true_values[j] - 3 * np.std(samples_no_prior_user[:, j]),
                        true_values[j] + 3 * np.std(samples_no_prior_user[:, j]))
            ax.set_ylim(true_values[i] - 3 * np.std(samples_no_prior_user[:, i]),
                        true_values[i] + 3 * np.std(samples_no_prior_user[:, i]))
        else:  # Upper triangle: blank
            ax.axis('off')

        # Add labels
        if j == 0:
            ax.set_ylabel(param_names[i], fontsize=8, labelpad=10)
        if i == n_params - 1:
            ax.set_xlabel(param_names[j], fontsize=8, labelpad=10)

# Add legend outside the plot
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', fontsize=10, ncol=2, frameon=False, bbox_to_anchor=(0.5, 1.05))

plt.tight_layout()
plt.show()
```

```python
# Step 1: Calculate maximum for each parameter
samples_prior_max = np.max(samples_prior_user, axis=0)
samples_no_prior_max = np.max(samples_no_prior_user, axis=0)

# Step 2: Compute max-based biases
bias_prior_max = samples_prior_max - true_values
bias_no_prior_max = samples_no_prior_max - true_values

# Step 3: Update the plotting code to include max-based biases
fig, axes = plt.subplots(n_params, n_params, figsize=(15, 15))

for i in range(n_params):
    for j in range(n_params):
        ax = axes[i, j]
        if i == j:  # Diagonal: 1D KDE plots
            kde_no_prior_user = gaussian_kde(samples_no_prior_user[:, i])
            kde_prior_user = gaussian_kde(samples_prior_user[:, i])
            x_vals = np.linspace(true_values[i] - 3 * np.std(samples_no_prior_user[:, i]),
                                 true_values[i] + 3 * np.std(samples_no_prior_user[:, i]), 300)
            ax.plot(x_vals, kde_no_prior_user(x_vals), color='blue', label='No Prior KDE', lw=1.5)
            ax.plot(x_vals, kde_prior_user(x_vals), color='green', label='With Prior KDE', lw=1.5)

            # True Value
            ax.axvline(true_values[i], color='red', linestyle='--', label='True Value')

            # Mean-based bias
            ax.axvline(samples_no_prior_mean[i], color='blue', linestyle='-', label=f'Mean Bias (No Prior: {bias_no_prior_mean[i]:.2e})')
            ax.axvline(samples_prior_mean[i], color='green', linestyle='-', label=f'Mean Bias (Prior: {bias_prior_mean[i]:.2e})')

            # Median-based bias
            ax.axvline(samples_no_prior_median[i], color='blue', linestyle=':', label=f'Median Bias (No Prior: {bias_no_prior_median[i]:.2e})')
            ax.axvline(samples_prior_median[i], color='green', linestyle=':', label=f'Median Bias (Prior: {bias_prior_median[i]:.2e})')

            # Max-based bias
            ax.axvline(samples_no_prior_max[i], color='blue', linestyle='-.', label=f'Max Bias (No Prior: {bias_no_prior_max[i]:.2e})')
            ax.axvline(samples_prior_max[i], color='green', linestyle='-.', label=f'Max Bias (Prior: {bias_prior_max[i]:.2e})')

            # Set x-limits and title
            ax.set_xlim(true_values[i] - 3 * np.std(samples_no_prior_user[:, i]),
                        true_values[i] + 3 * np.std(samples_no_prior_user[:, i]))
            ax.set_title(f"{param_names[i]}\nMean Bias: {bias_prior_mean[i]:.2e}\nMedian Bias: {bias_prior_median[i]:.2e}\nMax Bias: {bias_prior_max[i]:.2e}",
                         fontsize=10, pad=10)
        elif i > j:  # Lower triangle: 1-sigma ellipses
            reduced_samples_no_prior = samples_no_prior_user[:1000, [j, i]]
            reduced_samples_prior = samples_prior_user[:1000, [j, i]]

            kde_no_prior_user = gaussian_kde(reduced_samples_no_prior.T)
            kde_prior_user = gaussian_kde(reduced_samples_prior.T)

            x_vals = np.linspace(true_values[j] - 3 * np.std(samples_no_prior_user[:, j]),
                                 true_values[j] + 3 * np.std(samples_no_prior_user[:, j]), 100)
            y_vals = np.linspace(true_values[i] - 3 * np.std(samples_no_prior_user[:, i]),
                                 true_values[i] + 3 * np.std(samples_no_prior_user[:, i]), 100)
            X, Y = np.meshgrid(x_vals, y_vals)

            Z_no_prior_user = kde_no_prior_user(np.vstack([X.ravel(), Y.ravel()])).reshape(X.shape)
            Z_prior_user = kde_prior_user(np.vstack([X.ravel(), Y.ravel()])).reshape(X.shape)

            ax.contour(X, Y, Z_no_prior_user, colors='blue', linestyles='solid', levels=[np.max(Z_no_prior_user) * 0.606], alpha=0.7)
            ax.contour(X, Y, Z_prior_user, colors='green', linestyles='solid', levels=[np.max(Z_prior_user) * 0.606], alpha=0.7)

            # Add scatter for mean, median, and max positions
            ax.scatter(samples_no_prior_mean[j], samples_no_prior_mean[i], color='blue', s=40, label='Mean (No Prior)')
            ax.scatter(samples_prior_mean[j], samples_prior_mean[i], color='green', s=40, label='Mean (Prior)')
            ax.scatter(samples_no_prior_median[j], samples_no_prior_median[i], color='blue', s=40, marker='x', label='Median (No Prior)')
            ax.scatter(samples_prior_median[j], samples_prior_median[i], color='green', s=40, marker='x', label='Median (Prior)')
            ax.scatter(samples_no_prior_max[j], samples_no_prior_max[i], color='blue', s=40, marker='*', label='Max (No Prior)')
            ax.scatter(samples_prior_max[j], samples_prior_max[i], color='green', s=40, marker='*', label='Max (Prior)')

            # Add red lines for true values
            ax.axvline(true_values[j], color='red', linestyle='--', lw=1)
            ax.axhline(true_values[i], color='red', linestyle='--', lw=1)
            ax.set_xlim(true_values[j] - 3 * np.std(samples_no_prior_user[:, j]),
                        true_values[j] + 3 * np.std(samples_no_prior_user[:, j]))
            ax.set_ylim(true_values[i] - 3 * np.std(samples_no_prior_user[:, i]),
                        true_values[i] + 3 * np.std(samples_no_prior_user[:, i]))
        else:  # Upper triangle: blank
            ax.axis('off')

        # Add labels
        if j == 0:
            ax.set_ylabel(param_names[i], fontsize=8, labelpad=10)
        if i == n_params - 1:
            ax.set_xlabel(param_names[j], fontsize=8, labelpad=10)

# Add legend outside the plot
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', fontsize=10, ncol=3, frameon=False, bbox_to_anchor=(0.5, 1.05))

plt.tight_layout()
plt.show()
```
