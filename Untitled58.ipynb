```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Define covariance matrices
covariance_matrix = np.array([
    [6.59206e-7, -0.0000590403, 0.000241494, -0.551501, -40.6821, -1.25521e-6, -0.00191613],
    [-0.0000590403, 0.00396552, -0.0187743, 44.5582, 3298.6, 0.000100342, 0.145637],
    [0.000241494, -0.0187743, 0.0889077, -209.391, -15489.2, -0.000472575, -0.692714],
    [-0.551501, 44.5582, -209.391, 634261., 4.82757e7, 1.32679e7, 1403.59],
    [-40.6821, 3298.6, -15489.2, 4.82757e7, 3.69007e9, 100.021, 102061.],
    [-1.25521e-6, 0.000100342, -0.000472575, 1.32679e7, 100.021, 2.84274e-6, 0.00331947],
    [-0.00191613, 0.145637, -0.692714, 1403.59, 102061., 0.00331947, 5.81214]
])

covariance_matrix_prior = np.array([
    [3.0341e-7, -0.0000222726, 0.000106704, -0.149605, -10.1403, -4.10809e-7, -0.00100227],
    [-0.0000222726, 0.00165516, -0.00791195, 12.1292, 833.872, 0.0000322314, 0.0720583],
    [0.000106704, -0.00791195, 0.0378365, -56.9595, -3904.09, -0.000152401, -0.346713],
    [-0.149605, 12.1292, -56.9595, 175948., 1.34175e7, 0.366691, 376.301],
    [-10.1403, 833.872, -3904.09, 1.34175e7, 1.03865e9, 26.9674, 24025.9],
    [-4.10809e-7, 0.0000322314, -0.000152401, 0.366691, 26.9674, 8.27812e-7, 0.00115853],
    [-0.00100227, 0.0720583, -0.346713, 376.301, 24025.9, 0.00115853, 3.45927]
])

# Ensure covariance matrices are positive semi-definite
def make_positive_semi_definite(cov_matrix):
    """Ensure the covariance matrix is positive semi-definite."""
    eigvals, eigvecs = np.linalg.eigh(cov_matrix)
    eigvals[eigvals < 0] = 0  # Set negative eigenvalues to zero
    return eigvecs @ np.diag(eigvals) @ eigvecs.T

covariance_matrix_corrected = make_positive_semi_definite(covariance_matrix)
covariance_matrix_prior_corrected = make_positive_semi_definite(covariance_matrix_prior)

# Define true values and parameter names
true_values = np.array([8.33202e-6, 0.248521, 0.01, 20.5, 4.8, 0, 0])
param_names = [r"$M_c$", r"$\eta$", r"$\chi_1$", r"$\Lambda$", r"$\delta \Lambda$", r"$t_c$", r"$\phi_c$"]

# Function to plot ellipses
def plot_ellipse(ax, mean, cov, sigma=1.0, **kwargs):
    """Helper to plot ellipses."""
    vals, vecs = np.linalg.eigh(cov)
    vals[vals < 0] = 0  # Ensure no negative eigenvalues
    order = vals.argsort()[::-1]
    vals, vecs = vals[order], vecs[:, order]
    theta = np.degrees(np.arctan2(*vecs[:, 0][::-1]))
    width, height = 2 * sigma * np.sqrt(vals)
    ell = plt.matplotlib.patches.Ellipse(xy=mean, width=width, height=height, angle=theta, **kwargs)
    ell.set_facecolor('none')
    ax.add_patch(ell)

# Plotting
fig, axes = plt.subplots(len(true_values), len(true_values), figsize=(18, 18))
for i in range(len(true_values)):
    for j in range(len(true_values)):
        ax = axes[i, j]
        if i < j:
            ax.axis("off")
        elif i == j:
            # 1D Marginals
            x = np.linspace(
                true_values[i] - 3 * np.sqrt(covariance_matrix_corrected[i, i]),
                true_values[i] + 3 * np.sqrt(covariance_matrix_corrected[i, i]), 500
            )
            prior_pdf = multivariate_normal.pdf(x, mean=true_values[i], cov=covariance_matrix_prior_corrected[i, i])
            posterior_pdf = multivariate_normal.pdf(x, mean=true_values[i], cov=covariance_matrix_corrected[i, i])
            ax.plot(x, prior_pdf, color="blue", label="Prior")
            ax.plot(x, posterior_pdf, color="red", label="Posterior")

            # Median
            median = true_values[i]
            ax.axvline(median, color="green", linestyle="--", label="Median (True)")

            # Max value
            max_index = np.argmax(posterior_pdf)
            max_value = x[max_index]
            ax.axvline(max_value, color="orange", linestyle="--", label="Max")

            # Mean (True Value)
            ax.axvline(true_values[i], color="purple", linestyle="--", label="Mean")

            # Add text beside the plot in separate space
            ax.text(
                1.2, 0.5,
                f"Median: {median:.2e}\nMax: {max_value:.2e}\nMean: {true_values[i]:.2e}",
                ha="left", va="center", color="black", transform=ax.transAxes
            )

        else:
            # 2D Ellipses
            mean = [true_values[j], true_values[i]]
            cov_prior = covariance_matrix_prior_corrected[np.ix_([j, i], [j, i])]
            cov_post = covariance_matrix_corrected[np.ix_([j, i], [j, i])]
            ax.scatter(true_values[j], true_values[i], color="green", s=10)
            plot_ellipse(ax, mean, cov_prior, sigma=1, edgecolor="blue", label="Prior")
            plot_ellipse(ax, mean, cov_post, sigma=1, edgecolor="red", label="Posterior")

        if i == len(true_values) - 1:
            ax.set_xlabel(param_names[j])
        if j == 0:
            ax.set_ylabel(param_names[i])

# Single legend in one corner
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc="upper right", bbox_to_anchor=(1.2, 1.05))

plt.subplots_adjust(wspace=0.4, hspace=0.4)
plt.show()
```
