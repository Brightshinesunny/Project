```python
from scipy.stats import gaussian_kde
import numpy as np
import matplotlib.pyplot as plt

# η (대칭 질량 비율) 계산 함수
def symmetric_mass_ratio(m1, m2):
    return (m1 * m2) / (m1 + m2)**2

# \tilde{\Lambda} 계산
def tilde_lambda(m1, m2, lambda1, lambda2):
    eta = symmetric_mass_ratio(m1, m2)
    term1 = (1 + 7 * eta - 31 * eta**2) * (lambda1 + lambda2)
    term2 = np.sqrt(1 - 4 * eta) * (1 + 9 * eta - 11 * eta**2) * (lambda1 - lambda2)
    return (8 / 13) * (term1 + term2)

# \delta\Lambda 계산
def delta_lambda(m1, m2, lambda1, lambda2):
    eta = symmetric_mass_ratio(m1, m2)
    term1 = np.sqrt(1 - 4 * eta) * (1 - (13272 * eta) / 1319 + (8944 * eta**2) / 1319) * (lambda1 + lambda2)
    term2 = (1 - (15910 * eta) / 1319 + (32850 * eta**2) / 1319 + (3380 * eta**3) / 1319) * (lambda1 - lambda2)
    return 0.5 * (term1 + term2)

# 범위 및 샘플 수 설정
num_samples = 100000
m1_samples = np.random.uniform(0.5, 2.5, num_samples)
m2_samples = np.random.uniform(0.5, 2.5, num_samples)
lambda1_samples = np.random.uniform(0, 5000, num_samples)
lambda2_samples = np.random.uniform(0, 5000, num_samples)

# \tilde{\Lambda}와 \delta\Lambda 샘플 생성
tilde_lambda_samples = tilde_lambda(m1_samples, m2_samples, lambda1_samples, lambda2_samples)
delta_lambda_samples = delta_lambda(m1_samples, m2_samples, lambda1_samples, lambda2_samples)

# KDE 계산
kde_tilde_lambda = gaussian_kde(tilde_lambda_samples)
kde_delta_lambda = gaussian_kde(delta_lambda_samples)

# x축 범위 설정 (1%~99% 분위수)
tilde_lambda_range = np.linspace(
    np.percentile(tilde_lambda_samples, 1),
    np.percentile(tilde_lambda_samples, 99),
    1000
)
delta_lambda_range = np.linspace(
    np.percentile(delta_lambda_samples, 1),
    np.percentile(delta_lambda_samples, 99),
    1000
)

# PDF 계산
tilde_lambda_density = kde_tilde_lambda(tilde_lambda_range)
delta_lambda_density = kde_delta_lambda(delta_lambda_range)

# 개별 분포 시각화
plt.figure(figsize=(14, 6))

# \tilde{\Lambda} 분포
plt.subplot(1, 2, 1)
plt.plot(tilde_lambda_range, tilde_lambda_density, label=r'$\tilde{\Lambda}$ Density')
plt.fill_between(tilde_lambda_range, tilde_lambda_density, alpha=0.5)
plt.xlabel(r'$\tilde{\Lambda}$')
plt.ylabel('Density')
plt.title(r'Distribution of $\tilde{\Lambda}$')
plt.legend()

# \delta\Lambda 분포
plt.subplot(1, 2, 2)
plt.plot(delta_lambda_range, delta_lambda_density, label=r'$\delta\Lambda$ Density', color='orange')
plt.fill_between(delta_lambda_range, delta_lambda_density, alpha=0.5, color='orange')
plt.xlabel(r'$\delta\Lambda$')
plt.ylabel('Density')
plt.title(r'Distribution of $\delta\Lambda$')
plt.legend()

plt.tight_layout()
plt.show()
```
