```python
import numpy as np
import matplotlib.pyplot as plt

# Given true value and standard deviations
true_value = 0.2222
std_dev = 0.00173665
true_value_bias_1 = true_value +0.00648145  # Adding bias 1
std_dev_bias_1 = 0.00173901
true_value_bias_2 = true_value +0.0145833 # Adding bias 2
std_dev_bias_2 = 0.00174196

# Define the range for x (±3σ for all distributions)
x_min = min(true_value - 3 * std_dev, true_value_bias_1 - 3 * std_dev_bias_1, true_value_bias_2 - 3 * std_dev_bias_2)
x_max = max(true_value + 3 * std_dev, true_value_bias_1 + 3 * std_dev_bias_1, true_value_bias_2 + 3 * std_dev_bias_2)
x = np.linspace(x_min, x_max, 1000)

# Gaussian PDF formulas
pdf = (1 / (np.sqrt(2 * np.pi) * std_dev)) * np.exp(-((x - true_value) ** 2) / (2 * std_dev ** 2))
pdf_bias_1 = (1 / (np.sqrt(2 * np.pi) * std_dev_bias_1)) * np.exp(-((x - true_value_bias_1) ** 2) / (2 * std_dev_bias_1 ** 2))
pdf_bias_2 = (1 / (np.sqrt(2 * np.pi) * std_dev_bias_2)) * np.exp(-((x - true_value_bias_2) ** 2) / (2 * std_dev_bias_2 ** 2))

# Plot all PDFs with adjustments
plt.figure(figsize=(12, 7))
plt.plot(x, pdf, label=f"True Value: {true_value}, σ: {std_dev:.6f}")
plt.plot(x, pdf_bias_1, label=f"Biased True Value 1: {true_value_bias_1:.6f}, σ: {std_dev_bias_1:.6f}", linestyle='--')
plt.plot(x, pdf_bias_2, label=f"Biased True Value 2: {true_value_bias_2:.6f}, σ: {std_dev_bias_2:.6f}", linestyle=':')

# Adding vertical lines for the centers of distributions
plt.axvline(true_value, color='blue', linestyle=':', label=f"Center: {true_value}")
plt.axvline(true_value_bias_1, color='orange', linestyle=':', label=f"Center: {true_value_bias_1:.6f}")
plt.axvline(true_value_bias_2, color='green', linestyle=':', label=f"Center: {true_value_bias_2:.6f}")

# Adjust explanation box position
plt.title("Marginal PDF Comparison with Multiple Biases", loc='center')
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1))  # 그래프 외부 오른쪽 위에 배치

#plt.legend(loc='upper left')  # Move legend to top-left to avoid overlapping with curves
plt.grid()
plt.show()

# Analyze standard deviation differences
std_diff_1 = std_dev - std_dev_bias_1
std_diff_2 = std_dev - std_dev_bias_2
relative_diff_1 = (std_diff_1 / std_dev) * 100
relative_diff_2 = (std_diff_2 / std_dev) * 100

# Print differences
print("Standard Deviation Differences:")
print(f"True Value vs Biased True Value 1: Absolute Difference = {std_diff_1:.8f}, Relative Difference = {relative_diff_1:.2f}%")
print(f"True Value vs Biased True Value 2: Absolute Difference = {std_diff_2:.8f}, Relative Difference = {relative_diff_2:.2f}%")
```
