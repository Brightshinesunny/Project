```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. x축 정의 및 posterior PDF 생성 (모형 기반)
x = np.linspace(0, 1800, 2000)

# GW170817 스타일: 두 개의 피크를 가진 posterior 분포 생성
posterior_pdf = (
    3.2 * norm.pdf(x, loc=300, scale=100) +
    1.2 * norm.pdf(x, loc=600, scale=120)
)
posterior_pdf /= np.trapz(posterior_pdf, x)  # 정규화 (PDF 총합 = 1)

# 2. 각 EOS 위치 설정 (각각 고정된 Λ)
eos_names = ['WFF1', 'APR4', 'SLy', 'MPA1', 'H4', 'MS1b', 'MS1']
eos_lambdas = [200, 250, 300, 550, 950, 1400, 1550]  # 추정값

# 3. posterior에서 각 EOS 지점의 확률밀도 평가
posterior_values = np.interp(eos_lambdas, x, posterior_pdf)

# 4. Bayes Factor 계산 (WFF1 대비)
bayes_factors = posterior_values / posterior_values[0]

# 5. 시각화
plt.figure(figsize=(12, 6))

# posterior PDF 그리기
plt.plot(x, posterior_pdf, label='Posterior PDF (GW170817-like)', color='steelblue')
plt.fill_between(x, posterior_pdf, color='lightblue', alpha=0.4)

# 각 EOS마다 수직선 + 텍스트 (Bayes Factor 포함)
for name, lam, bf in zip(eos_names, eos_lambdas, bayes_factors):
    plt.axvline(x=lam, linestyle='--', color='red', alpha=0.6)
    plt.text(lam + 10, np.interp(lam, x, posterior_pdf) + 1e-5,
             f'{name}\nBF={bf:.2f}', rotation=90, va='bottom', ha='left', fontsize=9)

# 라벨 및 타이틀
plt.xlabel(r'$\tilde{\Lambda}$')
plt.ylabel('PDF')
plt.title('Posterior PDF with EOS Points and Bayes Factors (vs WFF1)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# 가상의 posterior 샘플 로드 또는 생성
# 각 waveform별 posterior 샘플이 있다고 가정
models = ['PhenomPNRT', 'PhenomDNRT', 'SEOBNRT', 'TaylorF2']
colors = ['royalblue', 'mediumseagreen', 'darkorange', 'orchid']
samples_dict = {
    'PhenomPNRT': np.random.normal(300, 80, 10000),
    'PhenomDNRT': np.random.normal(280, 90, 10000),
    'SEOBNRT':    np.random.normal(320, 100, 10000),
    'TaylorF2':   np.random.normal(400, 150, 10000),
}

# EOS 값
eos_names = ['WFF1', 'APR4', 'SLy', 'MPA1', 'H4', 'MS1b', 'MS1']
eos_lambdas = [200, 250, 300, 550, 950, 1400, 1550]

# x축 범위
x = np.linspace(0, 1800, 2000)

# prior
prior_pdf = np.full_like(x, 1 / (x.max() - x.min()))

# figure
fig, axs = plt.subplots(2, 1, figsize=(8, 10), sharex=True)

for ax in axs:
    # prior
    ax.plot(x, prior_pdf, label='Prior', color='gold')

    # 각 waveform model KDE
    for model, color in zip(models, colors):
        kde = gaussian_kde(samples_dict[model])
        pdf = kde(x)
        ax.plot(x, pdf, label=model, color=color)

    # EOS 수직선
    for name, lam in zip(eos_names, eos_lambdas):
        ax.axvline(x=lam, color='gray', lw=1)
        ax.text(lam, ax.get_ylim()[1]*0.95, name, ha='center', va='top', rotation=90)

    ax.set_ylabel('PDF')
    ax.set_ylim(0, 0.0035)
    ax.legend(fontsize=8)

axs[1].set_xlabel(r'$\tilde{\Lambda}$')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# 샘플 데이터 (이 부분은 실제 posterior 샘플로 교체해야 합니다)
np.random.seed(0)
Lambda_tilde_phenoPNRT = np.random.normal(300, 100, 10000)
Lambda_tilde_phenoDNRT = np.random.normal(310, 95, 10000)
Lambda_tilde_SEOBNRT = np.random.normal(320, 90, 10000)
Lambda_tilde_taylorF2 = np.random.normal(350, 110, 10000)

# Kernel Density Estimate
def kde_smooth(data):
    kde = gaussian_kde(data)
    x = np.linspace(0, 1600, 1000)
    y = kde(x)
    return x, y

plt.figure(figsize=(10, 6))

# Plot PDFs
for label, data, color in [
    ("PhenomPNRT", Lambda_tilde_phenoPNRT, 'tab:blue'),
    ("PhenomDNRT", Lambda_tilde_phenoDNRT, 'teal'),
    ("SEOBNRT", Lambda_tilde_SEOBNRT, 'darkorange'),
    ("TaylorF2", Lambda_tilde_taylorF2, 'orchid'),
]:
    x, y = kde_smooth(data)
    plt.plot(x, y, label=label, linewidth=2, color=color)
    if label == "PhenomPNRT":
        plt.fill_between(x, y, alpha=0.4, color=color)

# EOS vertical lines
eos_positions = {
    'WFF1': 200,
    'APR4': 250,
    'SLy': 300,
    'MPA1': 500,
    'H4': 800,
    'MS1b': 1350,
    'MS1': 1550
}

for eos, pos in eos_positions.items():
    plt.axvline(pos, color='gray', linewidth=1)
    plt.text(pos, 0.0035, eos, rotation=90, verticalalignment='bottom', horizontalalignment='center', fontsize=9)

plt.xlabel(r'$\tilde{\Lambda}$')
plt.ylabel('PDF')
plt.legend()
plt.tight_layout()
plt.show()
```
