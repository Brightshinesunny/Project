```python
import matplotlib.pyplot as plt
import numpy as np

# Define the fit function
def fit_function(ms):
    return (48466.1141*ms**7 - 511769.3262*ms**6 + 2290160.1775*ms**5 -
            5622239.7230*ms**4 + 8161248.7822*ms**3 - 6983220.6693*ms**2 +
            3244065.3811*ms - 624930.5178)

# Generate ms values and compute corresponding fit values
ms_values = np.linspace(1.2, 1.3, 300)
lambda_values = fit_function(ms_values)

# Biased points
biased_points = [
    (1.24805, 428.64),
    (1.25547, 304.45),
    (1.26510, 130.57)
]
true_point = (1.243, 528.28)

# Plot
fig, ax = plt.subplots(figsize=(5, 5))

# Plot the fitted curve
ax.plot(ms_values, lambda_values, label="Fit curve", color='blue', alpha=0.5)

# Plot biased points and true point
ax.scatter(*zip(*biased_points), color='black', label=r"$(\tilde{\Lambda}^{\rm inf}, m_s^{\rm inf})$")
ax.scatter(*true_point, color='red', label=r"$(\tilde{\Lambda}^{\rm true}, m_s^{\rm true})$")

# Axis settings
ax.set_xlim(1.2, 1.3)
ax.set_ylim(100, 1000)
ax.set_yscale("log")
ax.set_xticks([1.22, 1.24, 1.26, 1.28, 1.3])
ax.set_xlabel(r"$m_s\, (M_\odot)$", fontsize=14)
ax.set_ylabel(r"$\tilde{\Lambda}$", fontsize=14)

# Grid and formatting
ax.grid(True, which="both", linestyle="--", linewidth=0.5)
from matplotlib.ticker import LogFormatter
ax.yaxis.set_major_formatter(LogFormatter(base=10))

# Legend and layout
plt.legend()
plt.tight_layout()
plt.show()
```
