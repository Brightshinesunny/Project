```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import LogFormatter

# Define the APR4 fit function
def fit_function(ms):
    return (48466.1141 * ms**7
            - 511769.3262 * ms**6
            + 2290160.1775 * ms**5
            - 5622239.7230 * ms**4
            + 8161248.7822 * ms**3
            - 6983220.6693 * ms**2
            + 3244065.3811 * ms
            - 624930.5178)

# Generate APR4 fit curve
ms_values = np.linspace(1.3, 1.6, 300)
lambda_values = fit_function(ms_values)

# Points
biased_points = [(1.419, 443.8), (1.4425, 338.6), (1.478, 191.304)]  # MAP1 + ecc
true_point_map1 = (1.4, 528.28)  # MAP1 true
fit_true_point_apr4 = (1.4, 262.84)  # APR4 true
ecc_point_apr4 = (1.4127, 110.155)  # APR4 + ecc

# Plot
fig, ax = plt.subplots(figsize=(6, 6))

# Plot fit curve
ax.plot(ms_values, lambda_values, label="APR4 Fit Curve", color='blue', linestyle='--', alpha=0.7)

# Plot points with annotations
ax.scatter(*zip(*biased_points), color='black', label="MAP1 + eccentricity")
ax.scatter(*true_point_map1, color='red', label="MAP1 true value")
ax.scatter(*fit_true_point_apr4, color='blue', s=50, edgecolor='black', label="APR4 true value")
ax.scatter(*ecc_point_apr4, color='green', label="APR4 + eccentricity")

# Annotate each point
ax.annotate("MAP1 true", xy=true_point_map1, xytext=(1.405, 600), color='red', arrowprops=dict(arrowstyle="->", color='red'))
ax.annotate("APR4 true", xy=fit_true_point_apr4, xytext=(1.35, 300), color='blue', arrowprops=dict(arrowstyle="->", color='blue'))
ax.annotate("APR4 + ecc", xy=ecc_point_apr4, xytext=(1.36, 150), color='green', arrowprops=dict(arrowstyle="->", color='green'))
ax.annotate("MAP1 + ecc", xy=biased_points[0], xytext=(1.44, 500), color='black', arrowprops=dict(arrowstyle="->", color='black'))

# Axis settings
ax.set_xlim(1.3, 1.6)
ax.set_ylim(100, 1000)
ax.set_yscale("log")
ax.set_xticks([1.3, 1.4, 1.5])
ax.set_xlabel(r"$m_s\, (M_\odot)$", fontsize=14)
ax.set_ylabel(r"$\tilde{\Lambda}$", fontsize=14)

# Grid and formatting
ax.grid(True, which="both", linestyle="--", linewidth=0.5)
ax.yaxis.set_major_formatter(LogFormatter(base=10))

# Legend
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import LogFormatter

# Fit functions
def fit_apr4(ms):
    return (
        48466.1141 * ms**7 - 511769.3262 * ms**6 + 2290160.1775 * ms**5 -
        5622239.7230 * ms**4 + 8161248.7822 * ms**3 - 6983220.6693 * ms**2 +
        3244065.3811 * ms - 624930.5178
    )

def fit_mpa1(ms):
    log_lambda = -1.21 * ms**4 + 7.80 * ms**3 - 18.2 * ms**2 + 16.5 * ms - 1.46
    return 10 ** log_lambda

# Data points
map1_true = (1.4, 528.28)
map1_ecc = [(1.419, 443.8), (1.4425, 338.6), (1.478, 191.304)]
apr4_true = (1.4, fit_apr4(1.4))
apr4_ecc = [(1.4127, 110.155)]

# Prepare plot
fig, ax = plt.subplots(figsize=(6, 6))

# Plot fitted curves
ms_vals = np.linspace(1.3, 1.6, 300)
ax.plot(ms_vals, fit_apr4(ms_vals), label="APR4 fit", color="blue", linestyle="--")
ax.plot(ms_vals, fit_mpa1(ms_vals), label="MPA1 fit", color="gray", linestyle="-")

# Scatter points
ax.scatter(*map1_true, color='red', label="MAP1 true")
ax.scatter(*zip(*map1_ecc), color='black', label="MAP1 + $e_0$")
ax.scatter(*apr4_true, color='blue', label="APR4 true")
ax.scatter(*zip(*apr4_ecc), color='green', label="APR4 + $e_0$")

# Labels and axes
ax.set_xlim(1.3, 1.6)
ax.set_ylim(100, 1000)
ax.set_yscale("log")
ax.set_xticks([1.3, 1.4, 1.5])
ax.set_xlabel(r"$m_s\, (M_\odot)$", fontsize=14)
ax.set_ylabel(r"$\tilde{\Lambda}$", fontsize=14)
ax.yaxis.set_major_formatter(LogFormatter(base=10))
ax.grid(True, which="both", linestyle="--", linewidth=0.5)
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()
```
