```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import LogFormatter

# Fit functions
def fit_apr4(ms):
    return (
        48466.1141 * ms**7 - 511769.3262 * ms**6 + 2290160.1775 * ms**5 -
        5622239.7230 * ms**4 + 8161248.7822 * ms**3 - 6983220.6693 * ms**2 +
        3244065.3811 * ms - 624930.5178
    )

def fit_mpa1(ms):
    log_lambda = -1.21 * ms**4 + 7.80 * ms**3 - 18.2 * ms**2 + 16.5 * ms - 1.46
    return 10 ** log_lambda

# Data points
map1_true = (1.4, 528.28)
map1_ecc = [(1.419, 443.8), (1.4425, 338.6), (1.478, 191.304)]
apr4_true = (1.4, fit_apr4(1.4))
apr4_ecc = [(1.4127, 110.155)]

# Prepare plot
fig, ax = plt.subplots(figsize=(6, 6))

# Plot fitted curves
ms_vals = np.linspace(1.3, 1.6, 300)
ax.plot(ms_vals, fit_apr4(ms_vals), label="APR4 fit", color="blue", linestyle="--")
ax.plot(ms_vals, fit_mpa1(ms_vals), label="MPA1 fit", color="gray", linestyle="-")

# Scatter points
ax.scatter(*map1_true, color='red', label="MAP1 true")
ax.scatter(*zip(*map1_ecc), color='black', label="MAP1 + $e_0$")
ax.scatter(*apr4_true, color='blue', label="APR4 true")
ax.scatter(*zip(*apr4_ecc), color='green', label="APR4 + $e_0$")

# Labels and axes
ax.set_xlim(1.3, 1.6)
ax.set_ylim(100, 1000)
ax.set_yscale("log")
ax.set_xticks([1.3, 1.4, 1.5])
ax.set_xlabel(r"$m_s\, (M_\odot)$", fontsize=20)
ax.set_ylabel(r"$\tilde{\lambda}$", fontsize=20)
ax.yaxis.set_major_formatter(LogFormatter(base=10))
ax.grid(True, which="both", linestyle="--", linewidth=0.5)
ax.set_yscale("log")
ax.tick_params(axis='both', which='both', labelsize=19)  # 눈금 숫자 크게
ax.yaxis.set_major_formatter(LogFormatter(base=10))  # 로그 포맷 그대로

ax.tick_params(axis='both', which='major', labelsize=19)
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()
```
