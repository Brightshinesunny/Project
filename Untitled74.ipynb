```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd

# Times New Roman TTF 업로드해서 등록했다는 가정
font_path = "Times New Roman.ttf"
prop = fm.FontProperties(fname=font_path)

df = pd.read_csv("dmtotal_vs_e0_digitized.csv")

plt.figure(figsize=(6,4))

plt.scatter(df['e0'], df['Delta_Mtotal'], s=6, color='black')

# 축 라벨: Times New Roman, 크게
plt.xlabel(r"$\mathrm{e}_{0}$", fontsize=30, fontproperties=prop)
plt.ylabel(r"$\Delta \,\mathrm{M}_{tot}$", fontsize=30, fontproperties=prop)

ax = plt.gca()
ax.set_xticks([0.002, 0.006, 0.010])
ax.set_yticks([0.5, 1.0, 1.5])  # ✅ Y축 눈금을 원하는 값만 표시

# 눈금도 Times New Roman + 크게
for label in ax.get_xticklabels():
    label.set_fontproperties(prop)
    label.set_fontsize(30)

for label in ax.get_yticklabels():
    label.set_fontproperties(prop)
    label.set_fontsize(30)

plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("plot.pdf")  # LaTeX에 includegraphics로 삽입
plt.show()
```
