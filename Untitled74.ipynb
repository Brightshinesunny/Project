```python
# 런타임 최초 1회만 실행
!apt-get -qq update
!apt-get -qq install -y fonts-texgyre > /dev/null

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import os

# 사용 가능한 Times 계열 폰트 탐색
def pick_times_like_font(preferred_name="Times New Roman"):
    # 1) 사용자가 업로드한 TTF 중에 'Times New Roman'이 있으면 우선 사용
    user_fonts = [f for f in fm.findSystemFonts(fontpaths=['/content']) if 'Times' in os.path.basename(f)]
    for f in user_fonts:
        if 'Times New Roman' in os.path.basename(f) or 'TimesNewRoman' in os.path.basename(f):
            fm.fontManager.addfont(f)
            return fm.FontProperties(fname=f).get_name()

    # 2) 시스템에 'Times New Roman'이 이미 있으면 사용
    for f in fm.findSystemFonts():
        if 'Times New Roman' in f or 'TimesNewRoman' in f:
            fm.fontManager.addfont(f)
            return fm.FontProperties(fname=f).get_name()

    # 3) 대체: TeX Gyre Termes (apt로 설치된 Times 호환)
    for f in fm.findSystemFonts():
        if 'TeXGyreTermes' in f or 'texgyre' in f.lower():
            fm.fontManager.addfont(f)
            return fm.FontProperties(fname=f).get_name()

    # 4) 최후: 일반 serif
    return 'serif'

FONT_NAME = pick_times_like_font()
print("사용 폰트:", FONT_NAME)

# 전역 기본 폰트 설정
matplotlib.rcParams['font.family'] = FONT_NAME
matplotlib.rcParams['font.size'] = 25
```
