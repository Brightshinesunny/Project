```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib.ticker import LogFormatter

# --- Fit functions ---
def fit_apr4(ms):
    return (
        48466.1141 * ms**7 - 511769.3262 * ms**6 + 2290160.1775 * ms**5 -
        5622239.7230 * ms**4 + 8161248.7822 * ms**3 - 6983220.6693 * ms**2 +
        3244065.3811 * ms - 624930.5178
    )

def fit_mpa1(ms):
    log_lambda = -1.21 * ms**4 + 7.80 * ms**3 - 18.2 * ms**2 + 16.5 * ms - 1.46
    return 10 ** log_lambda

# --- 예시 데이터 ---
map1_true = (1.4, 528.28)
map1_ecc = [(1.419, 442.025), (1.44499, 334.556), (1.47998, 184.100)]
apr4_true = (1.4, fit_apr4(1.4))
apr4_ecc = [(1.4127, 110.155)]
map1_ecc2 = [(1.41923, 443.112), (1.44326, 337.002), (1.47691, 188.448)]
map1_ecc3 = (1.4914, 186.389)

# --- CSV 불러오기 ---
df = pd.read_csv("/content/APR4_EPP-m1eqm2-sample-100.csv")

# --- Times New Roman 폰트 등록 ---
font_path = "Times New Roman.ttf"   # Colab에 업로드한 경로
prop = fm.FontProperties(fname=font_path)

# --- Plot 준비 ---
fig, ax = plt.subplots(figsize=(11, 6))

ms_vals = np.linspace(1.3, 1.6, 300)
ax.plot(ms_vals, fit_apr4(ms_vals), label="APR4 EoS Fit", color="blue", linestyle="--")
ax.plot(ms_vals, fit_mpa1(ms_vals), label="MPA1 EoS Fit", color="gray", linestyle="-")

# --- 점 찍기 ---
ax.scatter(*map1_true, color='red', marker='^', s=128, label="MPA1 Reference Value")
ax.scatter(*apr4_true, color='blue', marker='H',s=136, label="APR4 Reference Value")

ax.scatter(*zip(*map1_ecc),
           color='black', marker='D', s=135,
           label=r"MPA1+$\mathrm{e}_{0}$(High-PN)")

ax.scatter(*zip(*map1_ecc2),
           color='yellow', marker='*', s=135,
           label=r"MPA1+$\mathrm{e}_{0}$(Low-PN)")

ax.scatter(*zip(*apr4_ecc),
           color='green', s=128,
           label="APR4 + $\mathrm{e}_{0}$")

# --- 축/라벨/옵션 ---
ax.set_xlim(1.3, 1.6)
ax.set_ylim(100, 1000)
ax.set_yscale("log")
ax.set_xticks([1.4, 1.5])

ax.set_xlabel(r"$\mathrm{m}_{1} \,/\, \mathrm{M}_{\odot}$", fontsize=32, fontproperties=prop)
ax.set_ylabel(r"$\tilde{\lambda}$", fontsize=32, fontproperties=prop)


# 눈금 Times New Roman 적용
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontproperties(prop)
    label.set_fontsize(36)

ax.yaxis.set_major_formatter(LogFormatter(base=10))
ax.grid(True, which="both", linestyle="--", linewidth=0.5)

# --- Legend Times New Roman 적용 ---
ax.legend(fontsize=36, loc="center left", bbox_to_anchor=(1.5, 0.8), prop=prop)

# y축 minor 라벨(중간 눈금)까지 Times New Roman 적용
for lab in ax.get_yminorticklabels():
    lab.set_fontproperties(prop)
    lab.set_fontsize(36)

# 범례 폰트/크기 확실히 적용
leg = ax.legend(fontsize=10, loc="center left", bbox_to_anchor=(1, 0.5), prop=prop)

for txt in leg.get_texts():
    txt.set_fontproperties(prop)
    txt.set_fontsize(30)


plt.tight_layout()
plt.show()
```
